```python
def extract_series_data(soup):
    """
    Извлекает информацию о сериале из HTML-разметки
    
    Args:
        html_content: HTML-строка с разметкой сериала
        
    Returns:
        Словарь с данными: title, production_year, views
    """
   
    result = {
        'title': None,
        'production_year': None,
        'views': None
    }
    
    title_element = soup.find('a', class_='short-title')
    if title_element:
        result['title'] = title_element.get_text(strip=True)
    
    
    for li in soup.find_all('li'):
        if 'Год выхода:' in li.get_text():
            year_match = re.search(r'\b(19|20)\d{2}\b', li.get_text())
            if year_match:
                result['production_year'] = year_match.group(0)
            break

    views_element = soup.find('div', class_='short-meta-item fx-1')
    if views_element:
        views_text = views_element.get_text(strip=True)
        views_match = re.search(r'\d+', views_text)
        if views_match:
            result['views'] = int(views_match.group(0))
    
    return result
```

```python
all_items = soup.find_all('div', class_='short-item')
all_items
```

```python
all_data = []

for item in all_items:
    all_data.append (extract_series_data(item))
all_df =pd.DataFrame( all_data)
all_df
```
