```python
prices = [int(i.get_text().replace(' ', ''))
          for i in soup.find_all('span', {'class' : "price_value"})]
```

```python
prices, names = [], []


for i in tqdm(range(1,101)):
    url = f'https://libroroom.ru/catalog/knigi/KHudozhestvennaya-literatura/?PAGEN_1={i}'
    page_res = requests.get(url).text
    soup = BeautifulSoup(page_res, 'html.parser')
    names += [i.get_text() for i in soup.find_all('div', {'class' : "item-title"})]
    prices += [int(i.get_text().replace(' ', ''))
          for i in soup.find_all('span', {'class' : "price_value"})]
```

# Дубль 2

```python
prices, names = [], []


for i in tqdm(range(1,101)):

    url = f'https://libroroom.ru/catalog/knigi/KHudozhestvennaya-literatura/?PAGEN_1={i}'
    page_res = requests.get(url).text
    soup = BeautifulSoup(page_res, 'html.parser')
    page_info = soup.find_all('div', {'class' : "catalog_item_wrapp grid-list__item grid-list-border-outer"})

    for element in page_info:
        names.append(element.find('div', {'class':"item-title"}).get_text())

        try:
            prices.append(int(element.find('span', {'class' : "price_value"}).get_text().replace(' ', '')))
        except:
            prices.append(-1)
```

```python
len(prices), len(names)
```

```python
df = pd.DataFrame()
```

```python
df['title'] = names
df['price'] = prices
```

```python
df
```

```python
df.to_csv('book_prices_dummy_site.csv', index = False)
```

# Средняя стоимость книги в ассортименте

```python
q = '''  
select avg(price) as avg_price, count(1) as cnt
from df
where price > 0
'''

duckdb.sql(q)
```

```python
q = '''  
select *
from df
limit 10
'''

duckdb.sql(q)
```

```python
q = '''  
select approx_quantile(price, 0.99)
from df
where price > 0 
'''

duckdb.sql(q).to_df()
```
