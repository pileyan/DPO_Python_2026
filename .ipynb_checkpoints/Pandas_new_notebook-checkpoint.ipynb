# JOIN


Посчитать, насколько отличается продолжительность композиции от средней по ее жанру

```python
duckdb.sql('''
with r as 
    (select track_genre, 
        avg(duration_ms) as avg_duration_ms
    from df
    group by 1)
    
select l.*,
    duration_ms - r.avg_duration_ms as diff
from df as l
join r
on l.track_genre = r.track_genre
order by duration_ms

''').to_df()
```

```python
duckdb.sql('''
    select track_genre, 
        avg(duration_ms) as avg_duration_ms
    from df
    group by 1
    order by avg(duration_ms)
''').to_df()
```

```python
aggregated_df = df.groupby('track_genre')['duration_ms'].mean().reset_index().sort_values(by = 'duration_ms')
```

```python
aggregated_df
```

```python
df = df.merge(aggregated_df, 
         how = 'inner', 
         left_on = 'track_genre', # [cписок полей]
         right_on = 'track_genre', # [cписок полей]
         suffixes = ('','_avg')
        )
```

```python
df['diff'] = df['duration_ms'] - df['duration_ms_avg']
df.sort_values(by = 'duration_ms')
```
