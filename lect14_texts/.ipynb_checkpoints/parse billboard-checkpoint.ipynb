# Download 1 year

```python
def one_year_billboard(year):
    
    link_new = f'https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_{year}'
    # Заголовки, скопированные из Firefox (F12 -> Сеть)
    firefox_headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:134.0) Gecko/20100101 Firefox/134.0',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
        'Accept-Language': 'ru-RU,ru;q=0.8,en-US;q=0.5,en;q=0.3',
        'Connection': 'keep-alive',
    }

    response = requests.get(link_new, headers=firefox_headers)
    bs = BeautifulSoup(response.text)
    table = bs.find_all('table')[0]
    columns = [i.get_text().strip() for i in table.find_all('th')]
    
    lst = []
    for i in range(0, len(table.find_all('tr')[1:])-1):
        lst.append([i.get_text().strip() 
                    for i in table.find_all('tr')[1:][i].find_all('td')])
    
    df = pd.DataFrame(lst)

    df.columns = columns

    return df
    
```

```python
from tqdm import tqdm

df = one_year_billboard(2010)
df['year'] = 2010

for i in tqdm(range(2011, 2026)):
    new_year_df = one_year_billboard(i)
    new_year_df['year'] = i
    df = pd.concat([df, new_year_df])
```

```python
def get_genres(artist_name):
    
    artist = artist_name.replace(' ', '_')
    artist_link = f'https://en.wikipedia.org/wiki/{artist}'
    
    result = requests.get(artist_link, headers=firefox_headers).text
    parsed = BeautifulSoup(result)
    try:

        soup = parsed.find_all('table')[0]

        genres_th = soup.find('th', string='Genres')
        if not genres_th:
            # Alternative: search by text content more flexibly
            genres_th = soup.find('th', text='Genres')

        # Get the parent <tr> and then the <td> with class "infobox-data"
        if genres_th:
            tr = genres_th.find_parent('tr')
            genres_td = tr.find('td', class_='infobox-data')
            if genres_td:
                # Find all <li> elements inside the <td>
                genre_items = genres_td.find_all('li')
                # Extract and clean text from each <li>
                genres = [li.get_text(strip=True) for li in genre_items]
#                 print(genres)
        else: 
            raise Exception('не сработало')
    except:
        soup = parsed.find_all('table')[1]

        genres_th = soup.find('th', string='Genres')
        if not genres_th:
            # Alternative: search by text content more flexibly
            genres_th = soup.find('th', text='Genres')

        # Get the parent <tr> and then the <td> with class "infobox-data"
        if genres_th:
            tr = genres_th.find_parent('tr')
            genres_td = tr.find('td', class_='infobox-data')
            if genres_td:
                # Find all <li> elements inside the <td>
                genre_items = genres_td.find_all('li')
                # Extract and clean text from each <li>
                genres = [li.get_text(strip=True) for li in genre_items]
    #             print(genres)
    
    return genres
```
