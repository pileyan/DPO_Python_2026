# JOIN


Посчитать, насколько отличается продолжительность композиции от средней по ее жанру

```python
duckdb.sql('''
with r as 
    (select track_genre, 
        avg(duration_ms) as avg_duration_ms
    from df
    group by 1)
    
select l.*,
    duration_ms - r.avg_duration_ms as diff
from df as l
join r
on l.track_genre = r.track_genre
order by duration_ms

''').to_df()
```

```python
duckdb.sql('''
    select track_genre, 
        avg(duration_ms) as avg_duration_ms
    from df
    group by 1
    order by avg(duration_ms)
''').to_df()
```

# Задания

1) В каких музыкальных жанрах наибольшая доля относительно популярных композиций (мы будем считать такой композицию , у которой popularity > 70) – решить нужно SQL'ем и pandas'ом :) – топ 5 вывести

```python
duckdb.sql('''
    select track_genre, 
        sum(case when popularity>70 then 1 else 0 end)::real/count(1) as popular_ratio
    from df
    group by 1
    order by 2 desc
    limit 5
''').to_df()
```

```python
df['is_popular'] = df['popularity'].apply(lambda x: 1 if x > 70 else 0)
aggregated = df.groupby('track_genre')['is_popular'].agg(['sum', 'count']).reset_index()
aggregated['popular_ratio'] = aggregated['sum']/aggregated['count']
aggregated.sort_values(by = 'popular_ratio', ascending = False).head(5)
```

2) Какова в зависимости от жанра средняя доля популярных песен на альбом

```python
duckdb.sql('''
    select track_genre,
        avg(popular_ratio) as avg_per_album_popular_ratio
    from
    (
        select track_genre,
            album_name, 
            sum(case when popularity>70 then 1 else 0 end)::real/count(1) as popular_ratio
        from df
        group by 1, 2
    )
    group by 1
    order by 2 desc
    limit 100
''').to_df()
```
