## 1. Самые наполненные альбомы. 

**Вопрос:** В каких альбомах больше всего треков? Вывести топ-10.

## 2. Артисты с наибольшим разнообразием жанров

**Вопрос:**
Какие артисты представлены в наибольшем количестве разных жанров? Топ-10.

## 3. Распределение популярности

**Вопрос:**
Разбейте popularity на бины (0–20, 20–40, …) и посчитайте количество треков в каждом.

 Использовать `pd.cut`

## 4. “Стабильные” артисты

**Вопрос:**
Найти артистов с **наименьшей дисперсией популярности** треков (минимум 20 треков).

**Метрика**: consistency = variance(popularity)

## 5. Энергия vs популярность

**Вопрос:**
Для каждого жанра посчитать корреляцию между `energy` и `popularity`.
Какие жанры имеют самую сильную положительную корреляцию?

## 6. Средний темп “хитовых” треков

**Вопрос:**
Сравнить средний `tempo`:

* у популярных (popularity > 70)
* у остальных

\+ Разница по жанрам.

## 7. “Танцевальные” жанры

**Вопрос:**
В каких жанрах максимальное среднее значение `danceability`, и при этом медианная популярность по жанру выше общей медианы

## 8. Альбомы-хиты

**Вопрос:**
Найти альбомы, где:

* минимум 5 треков
* не менее 50% треков имеют popularity > 60

Вывести топ-10 по средней популярности.

## 9. Длинные vs короткие треки

**Вопрос:**
Разделить треки на:

* короткие (< 3 минут)
* длинные (> 5 минут)

Сравнить:

* среднюю популярность
* среднюю energy

## 10. Кластеры жанров по характеристикам

**Вопрос:**
Сгруппировать жанры по средним значениям:

* danceability
* energy
* valence

И найти жанры, которые наиболее похожи друг на друга (например, через расстояние).


## 11. “Формула хита”

**Вопрос:**
Найти диапазоны признаков (примерно):

* danceability

в которых чаще всего встречаются треки с popularity > 75.
